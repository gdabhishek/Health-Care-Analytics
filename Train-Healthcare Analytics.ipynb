# XG Boost Hypertunning

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
# A parameter grid for XGBoost
params = {
    'max_depth':range(3,10,2),
     'min_child_weight':range(1,6,2),
     'gamma':[i/10.0 for i in range(0,5)],
        'reg_alpha':[1e-5, 1e-2, 0.1, 1, 100],
        'subsample':[i/100.0 for i in range(75,90,5)],
         'colsample_bytree':[i/100.0 for i in range(75,90,5)]
        }
```

```python
xgb = XGBClassifier(learning_rate=0.02, n_estimators=2000, objective='binary:logistic',
                    silent=True, nthread=1)
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
folds = 5
param_comb = 5

skf = StratifiedKFold(n_splits=folds, shuffle = True, random_state = 1001)
```

```python
random_search = RandomizedSearchCV(xgb, param_distributions=params, n_iter=param_comb, scoring='roc_auc', n_jobs=4, cv=skf.split(x,y), verbose=4, random_state=1001 )
```

```python
random_search.fit(x, y)
```

```python
random_search.best_score_
```

```python
random_search.best_params_
```
